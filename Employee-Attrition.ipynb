# Feature Engineering

```python
data.drop(['EmployeeCount', 'StandardHours', 'EmployeeNumber', 'Over18'], axis=1, inplace=True)
```

```python
data.shape
```

```python
def NoPromotion(data) :
    if data['YearsSinceLastPromotion'] == data['YearsAtCompany']:
        return 1
    else:
        return 0
data['NoPromotion'] = data.apply(lambda data:NoPromotion(data) ,axis = 1)
```

```python
data['PersonalLife'] = (data['WorkLifeBalance'] + data['RelationshipSatisfaction'])/2
```

```python
data['ApprovalRating'] = (data['PerformanceRating'] + data['JobInvolvement'])/2
```

```python
def LessPaidNotSatisfied(data):
    if data['JobSatisfaction'] < 3 and data['MonthlyIncome'] < 3500:
        return 1
    else:
        return 0
    
data['LessPaidNotSatisfied'] = data.apply(lambda data:LessPaidNotSatisfied(data), axis=1)
```

```python
data['TravelExpense'] = data['MonthlyIncome'] / data['DistanceFromHome']
```

```python
data['TimeAtEachCompany'] = (data['TotalWorkingYears'] - data['YearsAtCompany']) / data['NumCompaniesWorked']
data['TimeAtEachCompany'] = data['TimeAtEachCompany'].replace(np.Inf, 0)
```

```python
def LongEmployee(data):
    if data['YearsAtCompany'] > 8:
        return 1
    else: return 0
    
data['LongEmployee'] = data.apply(lambda data: LongEmployee(data), axis=1)
```

```python
data['Stability'] = data['YearsInCurrentRole'] / data['YearsAtCompany']
data['Stability'].fillna((data['Stability'].mean()), inplace=True)
```

```python
def YoungAndBadPaid(data):
    if data['Age'] < 34 and data['Age'] > 23 and data['MonthlyIncome'] < 3500:
        return 1
    else:
        return 0
    
data['YoungAndBadPaid'] = data.apply(lambda data: YoungAndBadPaid(data), axis=1)
```

```python
def Exhausted(data):
    if data['OverTime'] == 'Yes' and data['DistanceFromHome'] > 15:
        return 1
    else:
        return 0
    
data['Exhausted'] = data.apply(lambda data:Exhausted(data), axis=1)
```

```python
data.head()
```

```python
corr_matrix = data.corr()
corr_matrix['Attrition'].sort_values(ascending=False)
```

We can see that ***Exhausted*** has the highest positive correaltion with the target variable (***Attrition***) and ***TotalWorkingYears*** has the highest negative correlation. ***PerformanceRating*** has the least correaltion with Attrition.

```python
target = data['Attrition']

data.drop('Attrition', axis=1, inplace=True)
```

```python
# Splitting the data based on categorical and numerical features

num_features = []
for a in range(len(data.dtypes)):
    if data.dtypes[a] in ['int64', 'float64']:
        num_features.append(a)

data_num = data.iloc[:, num_features]
num_feature_names = data_num.columns

cat_features = data.columns.difference(data.columns[num_features])
data_cat = data.loc[:,cat_features]
```

```python
data_num.head()
```

```python
data_cat.head()
```

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler
```

```python
le = LabelEncoder()

for i in range(len(cat_features)):
    data_cat[cat_features[i]] = le.fit_transform(data_cat[cat_features[i]])
```

```python
data_cat.head()
```

```python
std_scaler = StandardScaler()

scaled = std_scaler.fit_transform(data_num)
scaled_data = pd.DataFrame(scaled, columns = num_feature_names)
```

```python
scaled_data.head()
```

```python
# Combining the numerical and categorical data

final_data = pd.concat([scaled_data, data_cat], axis=1)
final_data.shape
```

```python
# Removing Collinear Features

threshold = 0.8

corr_matrix = final_data.corr()
corr_matrix.head()

upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(np.bool))
upper.head()

to_drop = [column for column in upper.columns if any(upper[column] > threshold)]

final_data.drop(columns = to_drop, axis=1, inplace=True)

print('Columns to remove are:')

to_drop
```

# Fine-Tune the Model

```python
from sklearn.model_selection import cross_val_score
```

```python
def cross_val_metrics(model):
    scoring = ['accuracy', 'precision', 'recall']
    for sc in scoring:
        scores = cross_val_score(model, x_train, y_train, cv=5, scoring = sc)
        print(sc, ":", scores.mean())
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = [{'n_estimators':[500, 750, 1000],
             'max_depth':[3, 4, 5],
              'min_samples_leaf':[2, 3, 4]}]

gb_clf = GradientBoostingClassifier(random_state=42)

grid_search = GridSearchCV(gb_clf, param_grid, cv=5, scoring='accuracy')
grid_search.fit(x_train, y_train)
```
